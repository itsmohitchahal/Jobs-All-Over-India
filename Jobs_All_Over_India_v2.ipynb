# Analysing Jobs in the Legal Field
All jobs with any of following keywords: ["law", "attorney", "advocate" "trial", "lawyer", "legal", "law", "law-"]

```
df.describe()
```

```
law_keywords = ["law", "attorney", "advocate" "trial", "lawyer", "legal", "law-"]
# law_keywords = ["professor"]
law_df = pd.DataFrame(columns = ['title', 'tags', 'experience', 'salary', 'location']) 
# for ind in df.index: 
#   if ():
#      law_df.append(ind[ind])
# law_df = df.loc[any (s in df["title"] for s in law_keywords)]
for i in df.iterrows():
    if any (s in i[1].title.lower() for s in law_keywords):
      # print(i[1].title)
      law_df = law_df.append({'title' : i[1].title, 'tags' : i[1].tags, 'experience' : i[1].experience, 'salary': i[1].salary, 'location': i[1].location},  ignore_index = True)

# any(s in sentence for s in law_keywords)
```

```
law_df.describe()
```

```
law_job_title_data = law_df.location.str.strip().str.split(",", expand=True)[0].str.split(" ", expand=True)[0].value_counts().reset_index()
print("     For Jobs in the Legal Field")
print()
law_job_title_data.columns = ["Location", "Job Opportunities"]
law_job_title_data = law_job_title_data[:20]
law_job_title_data.style.background_gradient(cmap = "Reds")
```

```
law_job_title_data = law_df.title.str.strip().str.split(",", expand=True)[0].value_counts().reset_index()
law_job_title_data.columns = ["Job Title", "Job Opportunities"]
law_job_title_data = law_job_title_data[:20]
law_job_title_data.style.background_gradient(cmap = "Blues")
```

```
law_job_tags_data = law_df.tags.str.strip().str.split(",", expand=True)[0].value_counts().reset_index()
law_job_tags_data.columns = ["Tag", "Jobs with Requirement"]
law_job_tags_data = law_job_tags_data[:20]
law_job_tags_data.style.background_gradient(cmap = "Greens")
```
